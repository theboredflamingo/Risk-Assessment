```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
axis['Date'] = pd.to_datetime(axis['Date'])
axis.set_index('Date', inplace=True)

# Select relevant columns for forecasting
axisLSTM = axis[['Open', 'High', 'Low', 'Close', 'Volume', 'SMA_50', 'SMA_200', 'RSI', 'Prev Close']]
```

```python
scalers = {}
for column in axisLSTM.columns:
    scalers[column] = MinMaxScaler(feature_range=(0, 1))
    axisLSTM[column] = scalers[column].fit_transform(axisLSTM[[column]])
```

```python
train_size = int(len(axisLSTM) * 0.8)
train_data, test_data = axisLSTM[:train_size], axisLSTM[train_size:]
```

```python
def create_dataset(dataset, time_step=1):
    X, Y = [], []
    for i in range(len(dataset) - time_step - 1):
        a = dataset[i:(i + time_step)].values
        X.append(a)
        Y.append(dataset.iloc[i + time_step, 3])  # 'Close' is the target variable at index 3
    return np.array(X), np.array(Y)

time_step = 50
X_train, Y_train = create_dataset(train_data, time_step)
X_test, Y_test = create_dataset(test_data, time_step)

# Reshape input to be [samples, time steps, features] which is required for LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], X_test.shape[2])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()
model.add(Input(shape=(time_step, X_train.shape[2])))
model.add(LSTM(50, return_sequences=True))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
model
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model.fit(X_train, Y_train, batch_size=1, epochs=15, validation_split=0.2, callbacks=[early_stopping])
```

```python
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
train_predict = scalers['Close'].inverse_transform(train_predict)
test_predict = scalers['Close'].inverse_transform(test_predict)

# Invert the scaling for the actual values
Y_train = scalers['Close'].inverse_transform([Y_train])
Y_test = scalers['Close'].inverse_transform([Y_test])
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import datetime

# Define the stock ticker and download data
ticker = 'AXISBANK.NS'
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=365)
data = yf.download(ticker, start=start_date, end=end_date)

# Preprocessing
data['Date'] = data.index
data.reset_index(drop=True, inplace=True)

# Use only the 'Close' column
data_close = data[['Date', 'Close']]

# Create features (previous day's close price) and target (next day's close price)
data_close['Previous_Close'] = data_close['Close'].shift(1)
data_close = data_close.dropna()

# Split data into features and target
X = data_close[['Previous_Close']].values
y = data_close['Close'].values

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y.reshape(-1, 1))

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, shuffle=False)
```

```python
model2 = LinearRegression()
model2.fit(X_train, y_train)

# Predicting on test data
y_train_pred = model2.predict(X_train)
y_test_pred = model2.predict(X_test)

# Transform back to original scale
y_train_pred = scaler.inverse_transform(y_train_pred)
y_test_pred = scaler.inverse_transform(y_test_pred)
y_train_actual = scaler.inverse_transform(y_train)
y_test_actual = scaler.inverse_transform(y_test)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt
import datetime

def forecast(ticker):
    # Set dates for data download
    TODAY = datetime.datetime.today().date()
    Y1 = TODAY - datetime.timedelta(days=365)

    # Download data
    data = yf.download(ticker, start=Y1, end=TODAY)
    
    # Prepare data
    data['Previous_Close'] = data['Close'].shift(1)
    data = data.dropna()
    
    X = data[['Previous_Close']].values
    y = data['Close'].values
    
    # Scale data
    scaler = MinMaxScaler(feature_range=(0, 1))
    X_scaled = scaler.fit_transform(X)
    y_scaled = scaler.fit_transform(y.reshape(-1, 1))
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, shuffle=False)
    
    # Train model
    model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
    model.fit(X_train, y_train.ravel())
    
    # Predict using the test set
    y_pred = model.predict(X_test)
    
    # Inverse transform predictions and actual values to original scale
    y_test_inverse = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
    y_pred_inverse = scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()
    
    # Calculate MAE, MSE, and RMSE
    mae = mean_absolute_error(y_test_inverse, y_pred_inverse)
    mse = mean_squared_error(y_test_inverse, y_pred_inverse)
    rmse = sqrt(mse)
    
    # Print errors
    print("Mean Absolute Error (MAE):", mae)
    print("Mean Squared Error (MSE):", mse)
    print("Root Mean Squared Error (RMSE):", rmse)

# Example usage
forecast('AXISBANK.NS')
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import datetime
from math import sqrt

def predict(ticker):
    START = "2015-01-01"
    TODAY = datetime.date.today().strftime("%Y-%m-%d")
    
    # Download stock data
    data = yf.download(ticker, START, TODAY)
    
    # Prepare the data
    df_train = data[['Close']].copy()
    df_train.reset_index(inplace=True)
    df_train['Date'] = pd.to_datetime(df_train['Date'])
    df_train['Date_ordinal'] = df_train['Date'].map(pd.Timestamp.toordinal)

    if len(df_train) == 0:
        print('No valid data available for the selected stock. Please choose another stock or check your custom ticker.')
    else:
        X = df_train[['Date_ordinal']]
        y = df_train['Close']

        # Scaling the data
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        # Fit the model
        model = LinearRegression()
        model.fit(X_scaled, y)

        # Predict using the same dataset to calculate errors
        y_pred = model.predict(X_scaled)

        # Calculate MAE, MSE, and RMSE
        mae = mean_absolute_error(y, y_pred)
        mse = mean_squared_error(y, y_pred)
        rmse = sqrt(mse)
        
        # Print errors
        print("Mean Absolute Error (MAE):", mae)
        print("Mean Squared Error (MSE):", mse)
        print("Root Mean Squared Error (RMSE):", rmse)

# Example usage
predict('AXISBANK.NS')
```
